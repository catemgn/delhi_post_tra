```python
def reset_time_dim(ds):
    """
    Time dimension in WRF-Chem datasets is just the time index [0,1,2,3...].
    This function convert dimension Time as datetime[64].
    :param ds:
     wrf-chem xarray dataset.
    :type ds:xarray.Dataset.
    :return:
     xarray.Dataset with datetime time dim.
    :rtype: xarray.Dataset.
    """
    return ds.assign_coords(Time=pd.to_datetime(ds.XTIME.values))
```

```python
# reset time put in local time and discard 5 days spinup.
for k in d.keys():
    for v in d[k].keys():
        d[k][v]=reset_time_dim(d[k][v])  # reset
        d[k][v]=d[k][v].assign_coords(Time=pd.to_datetime(d[k][v].Time.to_index().tz_localize('UTC').tz_convert('Asia/Kolkata').strftime('%Y-%m-%d %H:%M:%S')))
        d[k][v].XTIME.values=d[k][v].Time.values
        d[k][v]=d[k][v].sel(Time=slice('2019-10-06T00:00:00.000000000','2019-10-14T23:30:00.000000000'))  # discad spinup
```

# Subset surface values of relevant pollutants

```python
ALT=d['BASE']['BASE'].ALT
def gas_ppm_to_ugm3(ds,gas,mol_w):
    """Convert gas from ppmv to ug/m3 based on molar weight of gas
    from  https://www.acom.ucar.edu/wrf-chem/ConversionMozartWRFChem.pdf"""
    ds[gas + "_ugm3"]=((ds[gas]/1e6)*(mol_w/29)*1e9)/ALT
    ds[gas + "_ugm3"].attrs["units"]="ug m-3"

# gas pollutants and their molar weight.
poll={"o3":48} 


for k in d.keys():
    for v in d[k].keys():
         for p,m in poll.items():
            gas_ppm_to_ugm3(d[k][v],p,m)
```

```python
# subset surface values of relevant pollutants.

def get_pollutants(ds):
    
    data ={ 
           "o3" : (['time',"lat","lon"], ds.o3_ugm3.values[:,0,:,:]),
           "pm25" : (['time',"lat","lon"], ds.PM2_5_DRY.values[:,0,:,:]),             
          }
    coords={'time': (['time'], ds.XTIME[:]),'lat': (['lat'], ds.XLAT[0,:,0]), 'lon': (['lon'], ds.XLONG[0,0,:])}
    p = xr.Dataset(data, coords=coords)
    return p
```

```python
mod={}
for k in d.keys():
    print(k)
    if k=='BASE':
        mod[k]={k: get_pollutants(d[k][k])}
    
    else:
        mod[k]={k2: get_pollutants(d[k][k2]) for k2 in d[k].keys()}
```

```python
pm25_metr={}

base=mod24['BASE']['BASE']

for k in mod24.keys():
    if k=='BASE':
        pm25_metr[k]={ k+'_vals': mod24[k][k].mean(['lat','lon','time']),
                 
                 }
    
    
    else:
        pm25_metr[k]={}
        for k2 in mod24[k].keys():
            pm25_metr[k].update({k2+'_vals': mod24[k][k2].mean(['lat','lon','time']),
                            k2+'_abs': (mod24[k][k2]-base).mean(['lat','lon','time']),
                            k2+'_%': ((mod24[k][k2]-base)/base*100).mean(['lat','lon','time'])})
```
