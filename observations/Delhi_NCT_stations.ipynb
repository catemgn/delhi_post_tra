This Notebook plots the stations of selected ground-based observations across Delhi NCT.

```python
# import modules
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
import pyproj
import pandas as pd
```

```python
# get NCT shapefile
```

```python
ncr_shp='../gadm_data/delhiNCT/delhiNCT.shp'
gdf = gpd.read_file(ncr_shp)
gdf.plot()
```

```python
# get ground- base sations corresponding to observations
```

```python
stations=pd.read_excel('cpcb_oct2019_cleaned.xlsx','stations')
```

```python
stations
```

```python
stations.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
stations.station.unique()
```

```python
#CONTEXTILY PLOTS

#DATA PREP
delhi_gp = gpd.GeoDataFrame.from_file(ncr_shp)

#Convert to local projection
transformer = pyproj.Transformer.from_crs("epsg:4326", delhi_gp.crs['init'])
hx = []
hy = []
for s, lat,lon in zip(stations['station'].unique(),stations['lat'].unique(),stations['lon'].unique()):
    xp, yp = transformer.transform(lat, lon)
    hx.append(xp)
    hy.append(yp)
```

```python
ax = delhi_gp.boundary.plot(color='k', linewidth=3, figsize=(8,8), label='NCT boundary', edgecolor='k')
plt.scatter(hy,hx,s=40,c='Blue')# 3
#plt.title('Delhi NCT ground-based stations for AQ evaluation', fontsize=20, fontname='Palatino Linotype', color='grey')
for i, txt in enumerate(stations['station'].unique()):
    ax.annotate(str(i+1), (stations['lon'].unique()[i], stations['lat'].unique()[i]),size=1,color='Black')
    
#Add in the basemap imagery
ax.tick_params(axis='both', which='major', labelsize=13)
plt.legend(fontsize=15)
plt.xlabel('LON',fontsize=13)
plt.ylabel('LAT',fontsize=13)
a=ctx.add_basemap(ax, crs=delhi_gp.crs['init'], source=ctx.providers.Esri.WorldTopoMap,zoom=15)
```
